### Irisデータセットで二値分類
(『Python機械学習プログラミング』p.27)

```python
'''
データセットの読み込み・前処理
'''
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

#Irisデータセットを読み込み
df = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data', header=None)

#1-100行目の目的変数の抽出
y = df.iloc[0:100, 4].values
#Iris-setosaを-1, Iris-virginicaを1に変換
y = np.where(y == 'Iris-setosa', -1, 1)
#1-100行目の1,3列目の抽出
X = df.iloc[0:100, [0,2]].values
```

```python
'''
データの可視化
'''
#品種setosaと品種versicolorのsepal lengthとpetal lengthをプロット
plt.scatter(X[:50,0], X[:50,1], color='red', marker='o', label='setosa')
plt.scatter(X[50:100,0], X[50:100,1], color='blue', marker='x', label='versicolor')
#軸のラベルの設定
plt.xlabel('sepal length [cm]')
plt.ylabel('petal length [cm]')
#凡例の設定
plt.legend(loc='upper left')
#図の表示
plt.show()
```

```python
'''
多層パーセプトロンモデルを定義
'''
import keras
from keras.layers.core import Dense, Activation
from keras.models import Sequential
from keras.optimizers import SGD
from keras.callbacks import EarlyStopping
from sklearn import datasets, preprocessing

#モデルを定義
early_stopping = EarlyStopping(monitor='val_loss', patience=2)
sgd = SGD(lr=0.001, decay=1e-6, momentum=0.9, nesterov=True)
model = Sequential()
model.add(Dense(0, input_dim=2))
model.add(Activation('relu'))
model.add(Dense(10, input_dim=500))
model.add(Activation('softmax'))
model.compile(loss='categorical_crossentropy', optimizer=sgd, metrics=["accuracy"])
```

## 手書き文字認識(多クラス分類)

```python
'''
MNISTデータを読み込み
'''
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import fetch_mldata

# MNISTデータをダウンロード・読み込み
mnist = fetch_mldata('MNIST original', data_home=".")

#X[i] : i番目の画像データ（28x28ピクセル）
X = mnist.data.astype('float32')
#y[i] : i番目の画像データのクラス（0-9）
y = mnist.target

#MNISTデータを表示(ランダムに25サンプルを描画)
p = np.random.randint(0, len(X)+1, 25)
for index, x in enumerate(X[p]):
    plt.subplot(5, 5, index + 1)
    plt.axis('off')
    plt.imshow(x.reshape(28, 28), cmap=plt.cm.gray_r, interpolation='nearest')
    plt.title('%i' % y[p][index])
plt.show()
```

```python
'''
前処理
'''
from sklearn.cross_validation import train_test_split
from sklearn.utils import shuffle

#データを正規化
X /= 255

#ラベルはバイナリ化
y = preprocessing.LabelBinarizer().fit_transform(y)

#シャッフル
X, y = shuffle(X, y, random_state=0)

#訓練データとテストデータに分割(訓練データ6万サンプル、テストデータ1万サンプル)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=10000, random_state=42)
```

```python
'''
多層パーセプトロンモデルを定義
'''
import keras
from keras.layers.core import Dense, Activation
from keras.models import Sequential
from keras.optimizers import SGD
from keras.callbacks import EarlyStopping
from sklearn import datasets, preprocessing

#モデルを定義
early_stopping = EarlyStopping(monitor='val_loss', patience=2)
sgd = SGD(lr=0.001, decay=1e-6, momentum=0.9, nesterov=True)
model = Sequential()
model.add(Dense(500, input_dim=784))
model.add(Activation('relu'))
model.add(Dense(10, input_dim=500))
model.add(Activation('softmax'))
model.compile(loss='categorical_crossentropy', optimizer=sgd, metrics=["accuracy"])
```

### モデルの構造

![モデル](model.png)

```python
'''
訓練・テスト実行
'''
#訓練
hist = model.fit(X_train, y_train, nb_epoch=20, batch_size=20, verbose=1,validation_split=0.1, callbacks=[early_stopping])

#テスト
score = model.evaluate(X_test, y_test)
```

```python
'''
訓練結果の表示
'''
#訓練履歴を取得
acc = hist.history['acc']
loss = hist.history['loss']
val_acc = hist.history['val_acc']
val_loss = hist.history['val_loss']

#訓練履歴をプロット
plt.plot(range(len(acc)), acc, label='acc')
plt.plot(range(len(loss)), loss, label='loss')
plt.plot(range(len(val_acc)), val_acc, label='var_acc')
plt.plot(range(len(val_loss)), val_loss, label='var_loss')
plt.legend()
plt.show()
```
