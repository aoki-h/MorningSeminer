# 活性化関数

- 入力の総和が一定値を超えると発火するニューロンをモデル化したもの。

- 入力に非線形性を与える役割。

- 活性化関数がなければ、いくら多層にしても単層パーセプトロンと同等モデルにしかならない。

**【活性化関数の種類】**<br>
**1. シグモイド関数**<br>
　閾値のあるニューロンの単純なモデルとしてよく使われていた。一定値以上になると飽和するため学習が進まなくなる場合がある。<br><br>
![シグモイド](activation_tanh.png)
 
**2. ReLU (Rectified Linear Unit)**<br>
　現在最もよく使われている活性化関数。値が飽和する問題を解消。計算量も少ないため学習が速い。<br><br>
![ReLU](activation_ReLU.png)

**3. Leaky ReLU**<br>
　ReLUで負の側の値も飽和しないように傾きを持たせたモデル。ReLUよりも性能が良いという報告がある。
![LeakyReLU](activation_LeakyReLU.png)

```python
import matplotlib.pyplot as plt
import numpy as np

#シグモイド
fig = plt.figure()
x = np.arange(-5,5,0.01)
plt.plot(x, np.tanh(x), linewidth=3)
plt.xlim(-5, 5)
plt.ylim(-1.1, 1.1)
plt.grid()
plt.savefig('activation_tanh.png')

#ReLU
fig = plt.figure()
plt.plot(x, np.maximum([0]*len(x), x), linewidth=3)
plt.xlim(-5, 5)
plt.ylim(-1.1, 1.1)
plt.grid()
plt.savefig('activation_ReLU.png')

#Leaky ReLU
fig = plt.figure()
plt.plot(x, np.maximum(0.01*x, x), linewidth=3)
plt.xlim(-5, 5)
plt.ylim(-1.1, 1.1)
plt.grid()
plt.savefig('activation_LeakyReLU.png')
```

## 手書き文字認識(多クラス分類)

28x28の手書き文字画像70000枚を含むデータセット(MNIST)を、多層パーセプトロンモデルで分類してみます。
- データを読み込む
- 訓練データとテストデータに分割したり、正規化したりなどの前処理
- 多層パーセプトロンモデルを定義
- モデルを訓練
- モデルをテスト
- 
コード全体は分からなくて大丈夫ですが、訓練の流れをおおまかに見ていきます。

```python
'''
MNISTデータを読み込み
'''
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import fetch_mldata

# MNISTデータをダウンロード・読み込み
mnist = fetch_mldata('MNIST original', data_home=".")

#X[i] : i番目の画像データ（28x28ピクセル）
X = mnist.data.astype('float32')
#y[i] : i番目の画像データのクラス（0-9）
y = mnist.target

#MNISTデータを表示(ランダムに25サンプルを描画)
p = np.random.randint(0, len(X)+1, 25)
for index, x in enumerate(X[p]):
    plt.subplot(5, 5, index + 1)
    plt.axis('off')
    plt.imshow(x.reshape(28, 28), cmap=plt.cm.gray_r, interpolation='nearest')
    plt.title('%i' % y[p][index])
plt.show()
```

```python
'''
前処理
'''
from sklearn.cross_validation import train_test_split
from sklearn.utils import shuffle
from sklearn import preprocessing

#データを正規化
X /= 255

#ラベルはバイナリ化
y = preprocessing.LabelBinarizer().fit_transform(y)

#シャッフル
X, y = shuffle(X, y, random_state=0)

#訓練データとテストデータに分割(訓練データ6万サンプル、テストデータ1万サンプル)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=10000, random_state=42)
```

```python
'''
多層パーセプトロンモデルを定義
'''
import keras
from keras.layers.core import Dense, Activation
from keras.models import Sequential
from keras.optimizers import SGD
from keras.callbacks import EarlyStopping
from sklearn import datasets

#モデルを定義
early_stopping = EarlyStopping(monitor='val_loss', patience=2)
sgd = SGD(lr=0.001, decay=1e-6, momentum=0.9, nesterov=True)
model = Sequential()
model.add(Dense(500, input_dim=784))
model.add(Activation('relu'))
model.add(Dense(10, input_dim=500))
model.add(Activation('softmax'))
model.compile(loss='categorical_crossentropy', optimizer=sgd, metrics=["accuracy"])

#モデルの構造を可視化してpngファイルに出力
from keras.utils.visualize_util import plot
plot(model, to_file='model.png')

#モデルの構造を表示
model.summary()
```

### モデルの構造

![モデル](model.png)

```python
'''
訓練
'''
#訓練実行
hist = model.fit(X_train, y_train, nb_epoch=20, batch_size=20, verbose=1,validation_split=0.1, callbacks=[early_stopping])
```

```python
'''
訓練結果の表示
'''
#訓練履歴を取得
acc = hist.history['acc']
loss = hist.history['loss']
val_acc = hist.history['val_acc']
val_loss = hist.history['val_loss']

#訓練履歴をプロット
plt.plot(range(len(acc)), acc, label='acc')
plt.plot(range(len(loss)), loss, label='loss')
plt.plot(range(len(val_acc)), val_acc, label='var_acc')
plt.plot(range(len(val_loss)), val_loss, label='var_loss')
plt.legend()
plt.show()
```

```python
'''
テスト実行
'''
#テスト
score = model.evaluate(X_test, y_test)
```
